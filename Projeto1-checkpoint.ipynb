```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline, BSpline
```

```python

growth = pd.read_excel("gdp_total_yearly_growth.xlsx")
exportacao = pd.read_excel("exports_percent_of_gdp.xlsx")
```

# Cressimento

Crescimento dos países da OECD

```python
paises_dentro = ['Canada','Turkey','Belgium','Luxembourg','Netherlands','Austria','Sweden','France','Portugal','Spain','Norway','Iceland','United Kingdom','United States']
growth_idx = growth.set_index('country')
growth_OECD = growth_idx.loc[paises_dentro,:]
growth_OECD
```

Crescimento de países parecidos com os da OECD

```python
paises_fora = ['China','Russia','Brazil','Mexico']
growth_general = growth_idx.loc[paises_fora,:]
growth_general
```

comparação entre o crescimento dos paises dentro e fora do grupo

```python
y = growth_OECD.mean()
y2 = growth_general.mean()
x = y.index
plt.figure(figsize=(20,5))
plt.title('Crescimento dos paises dentro X fora da OECD')
plt.xlabel('anos')
plt.ylabel('crescimento')
plt.plot(x,y,label="OECD")
plt.plot(x,y2,label="outros")
plt.legend()
```

deixando as linhas mais suaves

```python
xnew = np.linspace(x.min(),x.max(),1000)
spl = make_interp_spline(x, y, k=3)
spl_2 = make_interp_spline(x, y2, k=3)
power_smooth = spl(xnew)
power_smooth_2 = spl_2(xnew)
plt.figure(figsize=(20,5))
plt.title('Crescimento dos paises dentro X fora da OECD')
plt.xlabel('anos')
plt.ylabel('crescimento')
plt.plot(xnew,power_smooth,label="OECD")
plt.plot(xnew,power_smooth_2,label="outros")
plt.legend()
```
